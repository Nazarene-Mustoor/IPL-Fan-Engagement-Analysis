```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from pprint import pprint
import json
from selenium_stealth import stealth
usernames = ["jlo", "shakira", "beyonce", "katyperry"]
output = {}
def prepare_browser():
    chrome_options = webdriver.ChromeOptions()
    proxy = "server:port"
    chrome_options.add_argument(f'--proxy-server={proxy}')
    chrome_options.add_argument("start-maximized")
    chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
    chrome_options.add_experimental_option('useAutomationExtension', False)
    driver = webdriver.Chrome(options= chrome_options)
    stealth(driver,
        user_agent= 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.53 Safari/537.36',
        languages= ["en-US", "en"],
        vendor=  "Google Inc.",
        platform=  "Win32",
        webgl_vendor=  "Intel Inc.",
        renderer=  "Intel Iris OpenGL Engine",
        fix_hairline= False,
        run_on_insecure_origins= False,
        )
    return driver
def parse_data(username, user_data):
    captions = []
    if len(user_data['edge_owner_to_timeline_media']['edges']) > 0:
        for node in user_data['edge_owner_to_timeline_media']['edges']:
            if len(node['node']['edge_media_to_caption']['edges']) > 0:
                if node['node']['edge_media_to_caption']['edges'][0]['node']['text']:
                    captions.append(
                        node['node']['edge_media_to_caption']['edges'][0]['node']['text']
                    )

    output[username] = {
        'name': user_data['full_name'],
        'category': user_data['category_name'],
        'followers': user_data['edge_followed_by']['count'],
        'posts': captions,
    }
def scrape(username):
    url = f'https://instagram.com/{username}/?__a=1&__d=dis'
    chrome = prepare_browser()
    chrome.get(url)
    print (f"Attempting: {chrome.current_url}")
    if "login" in chrome.current_url:
        print ("Failed/ redir to login")
        chrome.quit()
    else:
        print ("Success")
        resp_body = chrome.find_element(By.TAG_NAME, "body").text
        data_json = json.loads(resp_body)
        user_data = data_json['graphql']['user']
        parse_data(username, user_data)
        chrome.quit()
def main():
    for username in usernames:
        scrape(username)
if __name__ == '__main__':
    main()
    pprint(output)
```

```python
from getpass import getpass
from time import sleep
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
#chrome browser
from selenium.webdriver import Chrome
import pandas as pd
from IPython.display import display, HTML

#Function to get tweet data
def get_tweet_data(card):
    """Extract data from tweet data"""
    username = card.find_element(By.XPATH, './/span').text
    try:
        postdate = card.find_element(By.XPATH,'.//time').get_attribute('datetime')
    except NoSuchElementException:
        return
    comment = card.find_element(By.XPATH, './/div[2]/div[2]/div[3]').text
    responding = card.find_element(By.XPATH, '//div[2]/div[2]/div[2]').text
    text = comment + responding
    reply_count = card.find_element(By.XPATH, '//button[@data-testid = "reply"]').text
    retweet_count = card.find_element(By.XPATH,'//button[@data-testid = "retweet"]').text
    like_count = card.find_element(By.XPATH,'//button[@data-testid = "like"]').text
    
    tweet = (username, postdate, text, reply_count, retweet_count, like_count)
    return tweet

#Start up the webdriver
driver = Chrome()

#Navigate to twitter and login
driver.get('https://www.x.com/login')
driver.maximize_window()
sleep(5)

# username = driver.find_element(By.XPATH,'//input[@name="text"]')
username = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//input[@name="text"]')))
username.send_keys('creamcheesenaz')
username.send_keys(Keys.RETURN)
sleep(3)

my_password = getpass()

# password = driver.find_element(By.XPATH,'//input[@name="password"]')
password = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//input[@name="password"]')))
password.send_keys(my_password)
password.send_keys(Keys.RETURN)
sleep(3)

#Find search input and search for hashtag
# search_input = driver.find_element(By.XPATH,'//input[@aria-label="Search query"]')
search_input = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//input[@aria-label="Search query"]')))
search_input.send_keys('#ipl2024')
search_input.send_keys(Keys.RETURN)
sleep(1)

#Navigate to 'Top' tab
# driver.find_element(By.LINK_TEXT, "Latest").click()
WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.LINK_TEXT, "Top"))).click()

#Get all tweets on the page
data = []
tweet_ids = set()
last_position = driver.execute_script("return window.pageYOffset;")
scrolling = True

scroll_attempts = 0
max_scroll_attempts = 50

while scrolling and scroll_attempts < max_scroll_attempts:
    page_cards = driver.find_elements(By.XPATH,'//article[@data-testid = "tweet"]')
    for card in page_cards[-30:]:  # Increase the number of processed tweets
        tweet = get_tweet_data(card)
        if tweet:
            tweet_id = ''.join(tweet)
            if tweet_id not in tweet_ids:
                tweet_ids.add(tweet_id)
                data.append(tweet)
                
    scroll_attempt = 0
    while True:
        # Check scroll position
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        sleep(3)  # Increase sleep time if needed
        curr_position = driver.execute_script("return window.pageYOffset;")
        if last_position == curr_position:
            scroll_attempt += 1
            
            # End of scroll region 
            if scroll_attempt >= 3:
                scrolling = False
                break
            else:
                sleep(3)  # Attempt to scroll again
        else:
            last_position = curr_position
            break
    scroll_attempts += 1
#Close the web driver
driver.close()

# Convert to DataFrame and print
df1 = pd.DataFrame(data, columns=['Username', 'TimeStamp', 'Text', 'Replies', 'Retweets', 'Likes'])
# Save DataFrame to CSV
df1.to_csv('ipl_2024_tweets.csv', index=False)
# Print DataFrame as HTML
html_table = df1.to_html()
display(HTML(html_table))
```

```python
import time
import random
from getpass import getpass
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException

# Function to get post data
def get_post_data(driver):
    try:
        caption = driver.find_element(By.XPATH, './/div[contains(@class, "C4VMK")]/span').text
    except NoSuchElementException:
        caption = ""
    try:
        likes = driver.find_element(By.XPATH, './/section[contains(@class, "EDfFK")]/div/div/button/span').text
    except NoSuchElementException:
        likes = "0"
    try:
        comments = driver.find_element(By.XPATH, './/li/div/div/div/div/span').text
    except NoSuchElementException:
        comments = "0"

    post_url = driver.current_url
    post = (caption, likes, comments, post_url)
    return post

# Start up the webdriver
options = Options()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36")
# Optional: Use proxy if needed
# options.add_argument('--proxy-server=http://your-proxy-server:port')  
driver = webdriver.Chrome(options=options)

# Navigate to Instagram and login
driver.get('https://www.instagram.com/accounts/login/')
driver.maximize_window()
time.sleep(5)

# Login
username = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.NAME, 'username')))
username.send_keys('martinmustoor')
password = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.NAME, 'password')))
password.send_keys(getpass())
password.send_keys(Keys.RETURN)
time.sleep(5)

# Navigate to hashtag page
driver.get('https://www.instagram.com/explore/tags/ipl2023/')
time.sleep(5)

# Get all posts and reels on the page
data = []
post_urls = set()
scroll_attempts = 0
max_scroll_attempts = 50

while len(post_urls) < 300 and scroll_attempts < max_scroll_attempts:
    try:
        cards = driver.find_elements(By.XPATH, '//a[contains(@href, "/p/") or contains(@href, "/reel/")]')
        for card in cards:
            try:
                post_url = card.get_attribute('href')
                if post_url and post_url.startswith('/'):
                    post_url = 'https://www.instagram.com' + post_url
                post_urls.add(post_url)
            except StaleElementReferenceException:
                continue
    except StaleElementReferenceException:
        continue

    scroll_attempts += 1
    # Scroll down to load more posts
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    time.sleep(random.uniform(3, 5))  # Random sleep to mimic human behavior

# Extract data from each post URL
data = []
for post_url in list(post_urls)[:300]:  # Limit to 300 posts
    try:
        driver.get(post_url)
        time.sleep(random.uniform(2, 4))  # Random sleep to mimic human behavior
        post_data = get_post_data(driver)
        if post_data:
            data.append(post_data)
    except Exception as e:
        print(f"Error processing {post_url}: {e}")
        continue

# Close the web driver
driver.quit()

# Convert to DataFrame and save
df = pd.DataFrame(data, columns=['Caption', 'Likes', 'Comments', 'URL'])
df.to_csv('ipl2023_instagram_posts.csv', index=False)
print(f"Scraped {len(df)} posts and reels")
```

```python
# Extract specific columns by index list
ipl_attendance = match_att[['Match Date','Teams','Venue']]

# Dictionary mapping venues to their capacities
venue_capacity = {
    "Dubai International Cricket Stadium": 25000,
    "Sharjah Cricket Stadium": 16000,
    "Narendra Modi Stadium, Ahmedabad": 132000,
    "Wankhede Stadium, Mumbai": 33000,
    "Brabourne Stadium, Mumbai": 50000,
    "Arun Jaitley Stadium, Delhi": 35200,
    "Rajiv Gandhi International Stadium, Uppal, Hyderabad": 55000,
    "M Chinnaswamy Stadium, Bengaluru": 40000,
    "Sawai Mansingh Stadium, Jaipur": 24000,
    "Eden Gardens, Kolkata": 66000,
    "MA Chidambaram Stadium, Chepauk, Chennai": 38200,
    "Maharaja Yadavindra Singh International Cricket Stadium, Mullanpur": 38000,
    "Bharat Ratna Shri Atal Bihari Vajpayee Ekana Cricket Stadium, Lucknow": 50100,
    "Barsapara Cricket Stadium, Guwahati": 37800,
    "Punjab Cricket Association IS Bindra Stadium, Mohali, Chandigarh": 27000,
}

# Create a new column for venue capacity
ipl_attendance["Capacity"] = ipl_attendance["Venue"].map(venue_capacity)


# Function to generate random attendance based on the venue capacity
def generate_random_attendance(capacity):
    # Generate random attendance between 65% to 100% of the venue capacity
    return random.randint(int(0.65 * capacity), capacity)

# Add Attendance column with random values
ipl_attendance["Attendance"] = ipl_attendance["Capacity"].apply(generate_random_attendance)

# Calculate the Attendance Percentage
ipl_attendance["Att %"] = (ipl_attendance["Attendance"] / ipl_attendance["Capacity"]) * 100
ipl_attendance["Att %"] = ipl_attendance["Att %"].clip(lower=65)  # Ensure minimum 65%
ipl_attendance["Att %"] = ipl_attendance["Att %"].map("{:.2f}%".format)  # Format with % symbol

# Remove "(20-over match)" from the "Match Date" column
ipl_attendance["Match Date"] = ipl_attendance["Match Date"].str.replace(" \(20-over match\)", "")


# Print DataFrame as HTML
html_table = ipl_attendance.to_html(index=False)
display(HTML(html_table))
```
